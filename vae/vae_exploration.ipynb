```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#statistics:
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import statsmodels.api as sm
from statsmodels.formula.api import ols

#chemistry:
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem import AllChem
from rdkit.Chem.MolStandardize import rdMolStandardize

import time as t
import random as r

print(f"Imported at: {t.ctime()}")
```

```python





def load_data(file_path):
    with open(file_path, 'r') as f:
        smiles = [line.strip() for line in f if line.strip()]
    return smiles

def clean_mol(molecule:str,iso:bool=False,stand:bool=False) -> Chem.Mol:
    
    try:
        mol = Chem.MolFromSmiles(molecule)
    
    except Exception as e:
        print(f"Error processing molecule: {molecule}")
        print(f"Error message: {e}")
        return None
    



    # get largest fragment
    mol = Chem.GetMolFrags(mol, asMols=True, sanitizeFrags=False)
    mol = max(mol, key=lambda m: m.GetNumAtoms())

    # decharge molecule
    
    uncharger = rdMolStandardize.Uncharger()
    mol = uncharger.uncharge(mol)
    

    #optional: sanitize molecule
    if stand:
        Chem.SanitizeMol(mol, sanitizeOps=Chem.SanitizeFlags.SANITIZE_KEKULIZE)
    

    return mol


#apply cleaning function to all molecules in dataset
mol = "CCO" #ethanole 
cleaned_mol = clean_mol(mol,stand=True)
print(Chem.MolToSmiles(cleaned_mol))
```
