```python
import numpy as np
import matplotlib.pyplot as plt
import time as t 
import random as r
import pandas as pd

import os 
import sys

print(f"imported at: {t.ctime()}")
```

```python
train_data_folder = "combination_1000_molecules_and_0_%_synthetic"
test_data_folder = "combination_500_molecules_and_0_%_synthetic"

def files_in_folder(folder):
    return [f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))]

train_files = files_in_folder(train_data_folder)
test_files = files_in_folder(test_data_folder)


def load_data(folder, files):
    data = []
    for f in files:
        file_path = os.path.join(folder, f)
        with open(file_path, 'r') as file:
            lines = file.readlines()
            for line in lines:
                data.append(line.strip())
    return data

def load_data_pd(folder, files):
    data = []
    for f in files:
        file_path = os.path.join(folder, f)
        df = pd.read_csv(file_path)
        data.append(df)
    return pd.concat(data, ignore_index=True)

train_data = load_data_pd(train_data_folder, train_files)
test_data = load_data_pd(test_data_folder, test_files)

print(train_data.info())
```
